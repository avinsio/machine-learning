## From Arrays to Images and Back

```python
from IPython.display import display
from PIL import Image
```

```python
im = Image.open("Chris.tiff")
display(im)
```

**Convert PIL Image to numpy array**

```python
import numpy as np
```

```python
array = np.array(im)
print(array.shape)
```

```python
array
```

**`uint8`** stands for **Unsigned 8-bit Integer**.

* **Unsigned** → Only positive values (no negative numbers).
* **8-bit** → Uses 8 bits of memory.
* **Range** → **0 to 255** (256 possible values).

In images:

* **0** = Black
* **255** = White
* **1–254** = Shades of gray


```python
mask = np.full(array.shape,225)
mask
```

```python
modified_array = array - mask

modified_array = modified_array * -1

modified_array = modified_array.astype(np.uint8)

modified_array
```

```python
display(Image.fromarray(modified_array))
```

```python
reshaped = np.reshape(modified_array,(100,400))
print(reshaped.shape)
```

```python
display(Image.fromarray(reshaped))
```
