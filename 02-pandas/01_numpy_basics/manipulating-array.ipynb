## Array Operations

```python
import numpy as np
```

```python
a = np.array([10,20,30,40])
b = np.array([1,2,3,4])
```

```python
c = a - b
print(c)
```

```python
d = a * b
print(d)
```

**convert farenheit to celcius**


```python
farenheit = np.array([0,-10,-5,-15,0])

celcius = (farenheit - 31) * (5/9)
celcius
```

```python
celcius > -20
```

```python
celcius % 2 ==0
```

**Elementwise manipulation**

```python
A = np.array([[1,1],[0,1]])
B = np.array([[2,0],[3,4]])
print(A*B)
```

**matrix product**

```python
print(A@B)
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
print(array1.dtype)

array2 =np.array([[7.1,8.2,9.1],[10.4, 11.2,12.3]])
print(array2.dtype)
```

```python
array3 =array1 + array2
print(array3)
print(array3.dtype)
```

Upcasting is the automatic conversion of a smaller data type to a larger or more precise data type.

**numpy have many aggregation function**

```python
print(array3.sum())
print(array3.max())
print(array3.min())
print(array3.mean())
```
