### Series
A Series in Pandas is a 1-dimensional labeled array.

```python
import pandas as pd
```

**Create Series**

```python
students = ['Alice', 'Jack', 'Molly']

pd.Series(students)
```

**Notes:**

* A **Series** is displayed with its values and index.
* Pandas automatically detects the data type (`dtype`).
* `object` dtype usually means **text/string** data.
* By default, the index starts from **0** and increases by **1**.


```python
numbers = [1,2,3,4]
pd.Series(numbers)
```

```python
students = ['Alice', 'Jack', None]
pd.Series(students)
```

**Notes:**

* None represents **missing data** in Python.
* Pandas handles None automatically.
* If a Series contains text + None, the **dtype becomes object**.
* Example: ['Alice', 'Jack', None] → dtype=object`.


```python
numbers = [1,2,3,None]
pd.Series(numbers)
```

**Notes:**

* NaN represents **missing numeric data**.
* NaN is different from None.
* If a numeric Series contains NaN, Pandas converts it to **float**.
* Reason: NaN is stored as a **floating-point value**.


```python
import numpy as np

np.nan ==None
```

**Notes:**

* None and NaN both represent **missing data**, but they are **different**.
* NaN != None


```python
np.nan == np.nan
```

```python
np.isnan(np.nan)
```

**Notes:**
* Use np.isnan() to check for NaN.
