### Querying a Series

**Notes:**

* Series can be accessed by **position** or **label**.
* `.iloc[]` → Access by **position** (0, 1, 2, ...).
* `.loc[]` → Access by **label** (index name).
* If no custom index is given, **position = label** (0, 1, 2, ...).


```python
import pandas as pd
```

```python
students_scores = {'Alice': 'Physics',
                   'Jack': 'Chemistry',
                   'Molly': 'English',
                    'Sam' :  'History'}
s = pd.Series(students_scores)
s
```

**Iloc attribute**

```python
s.iloc[3]
```

**loc attribute**

```python
s.loc["Molly"]
```

```python
s[3]
```

```python
s["Molly"]
```

```python
class_code = {99: 'Physics',
              100: 'Chemistry',
              101: 'English',
              102: 'History'}

s =pd.Series(class_code)
s
```

```python
s[0]
```

```python
s.iloc[0]
```
