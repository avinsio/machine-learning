### DataFrame

```python
import pandas as pd
```

```python
record1 = pd.Series({
    "Name" : "Alice",
    "class" : "Physics",
    "Score" : 85
})
record2 = pd.Series({
    "Name" : "Jack",
    "class" : "Chemistry",
    "Score" : 82
})
record3 = pd.Series({
    "Name" : "Helen",
    "class" : "Biology",
    "Score" : 90
})
```

**Create datafram**

```python
df =pd.DataFrame([record1,record2,record2],
                 index=["school1","school2","school1"])
```

```python
df.head()
```

**Another way to create dataframe**

```python
students= [{
    "Name" : "Alice",
    "class" : "Physics",
    "Score" : 85},
    {
    "Name" : "Jack",
    "class" : "Chemistry",
    "Score" : 82
    },
    {
    "Name" : "Helen",
    "class" : "Biology",
    "Score" : 90   
    }

]

df =pd.DataFrame(students,index=["school1","school2","school1"])
```
