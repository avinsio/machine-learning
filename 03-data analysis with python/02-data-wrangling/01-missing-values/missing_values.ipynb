```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Student": ["Avi", "Neha", "Rahul", "Riya", "John",
                "Sara", "Aman", "Priya", "Karan", "Meera"],

    "Age": [22, np.nan, 24, 23, np.nan,
            21, 22, np.nan, 24, 23],

    "Gender": ["Male", "Female", np.nan, "Female", "Male",
               "Female", "Male", np.nan, "Male", "Female"],

    "Marks": [88, 91, np.nan, 79, 85,
              np.nan, 92, 87, np.nan, 90],

    "City": ["Delhi", "Mumbai", "Delhi", np.nan, "Pune",
             "Delhi", "Mumbai", "Pune", np.nan, "Delhi"],

    "Salary": [45000, 52000, np.nan, 48000, 51000,
               np.nan, 56000, 49000, 47000, np.nan]
})

df
```

### Detect Missing Values

**Display the dataset.**

```python
df.head()
```

**Check if there are any missing values.**

```python
df.isnull().values.any()
```

**Count missing values in each column**

```python
df.isna().sum()
```

**Count total missing values**

```python
df.isna().sum().sum()
```

**Find the percentage of missing values in each column.**

```python
df.isna().mean() *100
```

**Show only rows containing at least one missing value.**

```python
df[df.isna().any(axis=1)]
```

**Show rows with no missing values.**

```python
df[df.notnull().all(axis=1)]
```

### Drop Missing Values

**Drop every row containing at least one missing value**

```python
df.dropna(axis=0)
```

**Drop every column containing at least one missing value.**

```python
df.dropna(axis=1)
```

**Drop rows only if all values are missing.**

```python
df.dropna(how="all")
```

**Keep only rows having at least 4 non-null values**

```python
df.dropna(thresh=4)
```

**Drop columns having more than 2 missing values**

```python
df.dropna(axis=1, thresh = len(df)-2)
```

### Replace Missing Values

**Replace missing Age with mean.**

```python
import numpy as np
```

```python
mean= df["Age"].mean()
print(mean)
df["Age"].replace(np.nan, mean)
```

**Replace missing Marks with median.**

```python
median = df["Marks"].median()
df["Marks"].replace(np.nan,median)
```

**Replace missing Salary with mean.**

```python
mean_salary = df["Salary"].mean()
print(mean_salary)
df["Salary"].replace(np.nan, mean_salary)
```

**Replace missing Gender with mode.**

```python
mode = df["Gender"].mode()[0]
df["Gender"].replace(np.nan, mode)
```

**Replace missing City with mode.**

```python
mode= df["City"].mode()[0]
df["City"].fillna(mode)
```

**Replace every NaN with "Unknown".**

```python
df.replace(np.nan,"Unknown")
```

**Replace only Salary NaNs with 0.**

```python
df["Salary"].fillna(0)
```

**Replace Age NaNs using forward fill.**

```python
df["Age"].ffill()
```

**Replace Marks NaNs using backward fill**

```python
df["Marks"].bfill()
```
