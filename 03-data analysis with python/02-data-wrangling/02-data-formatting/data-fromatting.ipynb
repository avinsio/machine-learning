### Text Formatting

**Convert every city name to lowercase.**

```python
df.City = df.City.str.lower()
```

**Convert Gender into title case.**

```python
df.Gender= df.Gender.str.title()
```

**Count unique city names before cleaning.**

```python
df.City.nunique()
```

**Standardize**
- NY
- N.Y.
- new york

into

- New York

```python
df.City = df.City.replace(["new york", "n.y.","ny"],"New York" )
```

**Count unique cities again.**

```python
df.City.nunique()
```

**Data Types**

**Check the data type of Price.**

```python
df["Price"].dtypes
```

**Convert Price from object to integer.**

```python
df["Price"] =pd.to_numeric(df["Price"])
df["Price"]
```

**Verify the conversion.**

**Calculate**
