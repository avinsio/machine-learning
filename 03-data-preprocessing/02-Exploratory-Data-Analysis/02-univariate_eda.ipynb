```python
import pandas as pd
import seaborn as sns
```

```python
df =pd.read_csv("dataset/train.csv")
```

```python
df.info()
```

***1. Categorical data***

**a. Countplot**

* **Purpose:** Shows the **frequency (count)** of each category in a categorical feature.

**Questions to ask during EDA:**

* Which category is the **most common**?
* Which category is the **least common**?
* Is the data **balanced** across categories?
* Are there any **rare or underrepresented** categories?
* Are there any **unexpected, inconsistent, or misspelled** categories ?
* Do any categories require **data cleaning or grouping**?


```python
sns.countplot(x='Survived', data=df)
df["Survived"].value_counts()
```

```python
df["Survived"].value_counts().plot(kind="bar");
```

```python
# Pclass
sns.countplot(x="Pclass", data=df)
df["Pclass"].value_counts()
```

**b. PieChart**
* Shows the percentage/proportion of each category.
* Best for showing parts of a whole.
* Use only with a small number of categories

```python
df['Survived'].value_counts().plot(kind="pie", autopct='%.2f');
```

```python
df['Sex'].value_counts().plot(kind="pie", autopct='%.2f');
```

**2. Numerical data**

**Histogram**

Purpose

Shows the distribution of a numerical feature by grouping values into bins.

Why do we use it?
* Understand how values are distributed.
* Identify skewness.
* Detect possible outliers.
* Check whether data is approximately normal.

Questions to ask
* Which range has the most observations?
* Is the distribution normal?
* Is it left or right skewed?
* Are there multiple peaks?
* Are there gaps?
* Are there possible outliers?

```python
import matplotlib.pyplot as plt
plt.hist(df["Age"],bins=20);
```

```python
sns.histplot(df["Age"])
```

**b. histplot**

kde=True: Adds a smooth curve to better visualize the distribution's shape.

```python
sns.histplot(df["Age"], kde=True)
```

```python
sns.histplot(df["Age"], kde=True, stat="density")
```

**c. Boxplot**

Purpose

Shows the spread (distribution) of a numerical feature and helps identify outliers using the five-number summary.

Why do we use it?
* Understand the spread of the data.
* Detect outliers.
* Find the median.
* Measure variability using the IQR.
* Compare distributions across groups (in bivariate analysis).

Questions to ask during EDA
* What is the median value?
* Are there any outliers?
* Is the data skewed?
* How spread out is the data (IQR)?
* Is there high or low variability?

```python
sns.boxplot(x=df["Fare"])
```

```python
sns.boxplot(x=df["Age"])
```
